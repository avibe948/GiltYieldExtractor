```python
import pandas as pd

url = "https://www.hl.co.uk/shares/corporate-bonds-gilts/bond-prices/uk-gilts?column=coupon&order=desc"

bond_data = pd.read_html(url)[0]

bond_data['ShortName'] = bond_data['Issuer'].apply(lambda x: x.split('|')[1].strip())
bond_data["Price"] = bond_data["Price"]
bond_data["Maturity"] = pd.to_datetime(bond_data["Maturity"], format="%d %B %Y")
bond_data['Ttm'] = (bond_data['Maturity'] - pd.to_datetime('today')).dt.days / 365
bond_data['CouponFreq']=2
bond_data=bond_data.rename(columns={'Coupon (%)': 'Coupon'})
bond_data=bond_data.drop(columns=['Issuer','Actions'])
col = bond_data.pop('ShortName')
bond_data.insert(0, 'ShortName', col)

print(bond_data.head())
```

```python
FACE_VALUE=100.0
```

```python
import numpy_financial as npf
def calculate_ytm(coupon, face_value, years_to_maturity, price):
    coupon = coupon / 100.0
    periods = years_to_maturity
    ytm = npf.rate(nper=periods, pmt=coupon*face_value/periods, pv=-price, fv=face_value)
    return ytm * 100.0

def calculate_dirty_price(coupon:float, face_value:float, years_to_maturity:float, clean_price:float, payment_frequency:int):
    coupon = coupon / 100.0
    periods = int(years_to_maturity * payment_frequency)
    days_in_period = int(365 / payment_frequency)
    days_since_last_coupon = datetime.datetime.now().timetuple().tm_yday % days_in_period
    coupon_payment = (face_value * coupon) / payment_frequency
    accrued_interest = coupon_payment * days_since_last_coupon / days_in_period
    dirty_price = clean_price + accrued_interest
    return dirty_price
```

```python
def calc_simple(coupon):
    return coupon + 1
```

```python
bond_data.head(1)
```

```python
bond_data['DirtyPrice']=bond_data.apply(lambda x: calculate_dirty_price(x['Coupon'],FACE_VALUE,x['Ttm'],x['Price'],x['CouponFreq']),axis=1)
```

```python
bond_data.dtypes
```

```python
bond_data['Yield'] = bond_data.apply(lambda x: calculate_ytm(x['Coupon'], 100, x['Ttm'], x['DirtyPrice']), axis=1)
```

```python
bond_data=bond_data.sort_values('Ttm')
```
